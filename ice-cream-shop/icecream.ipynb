# Question 3 Part A)

```python
monthly_info = pd.read_csv('Monthly Ice Cream Shop Sales by Flavor.csv')
monthly_info
```

```python
sns.lineplot(monthly_info, x='Month', y='Total Units')
plt.xlabel('Month')
plt.ylabel('Total Units')
plt.show()
```

# Question 3 Part A Optional)

```python
# Creating the different flavors on a line chart
plt.plot(monthly_info['Month'], monthly_info['Vanilla'], label='Vanilla')
plt.plot(monthly_info['Month'], monthly_info['Chocolate'], label='Chocolate')
plt.plot(monthly_info['Month'], monthly_info['Strawberry'], label='Strawberry')
plt.plot(monthly_info['Month'], monthly_info['Neapolitan'], label='Neapolitan')
plt.plot(monthly_info['Month'], monthly_info['Mango'], label='Mango')
plt.plot(monthly_info['Month'], monthly_info['Lime'], label='Lime')

# labeling the chart and moving legend outside of chart
plt.xlabel('Months')
plt.ylabel('Units Sold')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Sales by Flavor')
plt.show()
```

# Question 3 Part B)

```python
# creating the initial bar chart
plt.bar(monthly_info['Month'], monthly_info['Total Profit'])

# labeling the chart
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.title('Monthly Total Profit')
plt.show()
```
