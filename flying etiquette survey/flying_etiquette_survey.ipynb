# Clean the Data

```python
# check for duplicates
df.duplicated().sum()
```

```python
# check for null values
df.isna().sum()
```

```python
# drop all nulls in column Age
df.dropna(inplace=True)
```

```python
df['How tall are you?']
```

```python
# adds a new column based on the "How tall are you column"
df['Height (Inches)'] = df['How tall are you?']

# changes hights for anything over 6ft 6inch or under 5ft

def convert_heights(height):
    if height == "Under 5 ft.":
        return 60
    elif height == '6\'6" and above':
        return 78
    else:
        feet, inches = height.split("'")
        inches = inches.replace('"', '')
        return int(feet) * 12 + int(inches)    


```

```python
# change Height column to inches
df['Height (Inches)'] = df['Height (Inches)'].apply(convert_heights)
```

# Visuals

```python
plt.figure(figsize=(8,5), dpi=80)

sns.countplot(df, x='Age', order=['18-29', '30-44', '45-60', '> 60'], stat='percent')
plt.title("Percent of participants by age")
plt.show()
```

```python
plt.figure(figsize=(8,5), dpi=100)
sns.histplot(df, x='Is it rude to ask someone to switch seats with you in order to be closer to friends?', stat='percent')
plt.xlabel('Survey Response')
```

```python
sns.displot(df, x='Is it rude to ask someone to switch seats with you in order to be closer to friends?',
           stat='percent',
           col='Age',
           aspect=1.5,
           height=5,
           col_order=['18-29', '30-44', '45-60', '> 60'],
           col_wrap=2
           )
```
